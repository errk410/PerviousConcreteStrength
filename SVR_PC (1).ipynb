Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVR
```

```python
# svr=SVR(kernel='linear')
```

```python
svr=SVR(kernel='rbf')
```

```python
svr.fit(X_train,y_train)
```

```python
svr.coef_
```

```python
y_train_pred=svr.predict(X_train)
```

```python
y_pred=svr.predict(X_test)
```

```python
#Model Performance on Training Data

svr.score(X_train, y_train)

round(svr.score(X_train, y_train)*100)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_pred))
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_absolute_error
```

```python
error_score = metrics.r2_score(y_train, y_train_pred)
mae = metrics.mean_absolute_error(y_train, y_train_pred)
mse = mean_squared_error(y_train, y_train_pred)
rmse = np.sqrt(mse)
print("R squared Error : ", error_score)
print("Mean Absolute Error : ", mae)
print('MSE = ', mse)
print('RMSE = ', rmse)
```

```python
error_score = metrics.r2_score(y_test, y_pred)
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("R squared Error : ", error_score)
print("Mean Absolute Error : ", mae)
print('MSE = ', mse)
print('RMSE = ', rmse)
```

```python
plt.scatter(y_train,y_train_pred)
plt.xlabel("Train strength")
plt.ylabel("Predicted strength")
plt.title(" Train strength vs Predicted strength")
plt.show()
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("Test strength")
plt.ylabel("Predicted strength")
plt.title("Test strength vs Predicted strength")
plt.show()
```
